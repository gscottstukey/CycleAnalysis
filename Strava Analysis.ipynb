```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from StravaAnalyzer import StravaAnalyzer
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.cross_validation import train_test_split
import seaborn as sns
%matplotlib inline
```

```python
s = StravaAnalyzer()
s.fill_df()
```

```python
df = pd.read_csv('hills.csv')
df = df.drop(['Unnamed: 0', 'distance'], 1)
df.head()
```

```python
lin = LinearRegression()
X = df[['previous_climb', 'previous_distance', 'score', 'time_riding']].values
y = df.pop('velocity')
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
lin.fit(X_train, y_train)
lin.score(X_test, y_test)
```

```python
fig, axs = plt.subplots(2,2, figsize=(12,8))
axs[0][0].scatter(df['score'], y)
axs[0][1].scatter(df['previous_climb'], y)
axs[1][0].scatter(df['previous_distance'], y)
axs[1][1].scatter(df['time_riding'], y)
```

```python
rf = RandomForestRegressor()
rf.fit(X, y)
sorted(zip(rf.feature_importances_, ['previous_climb', 'previous_distance', 'score', 'time_riding']),reverse=True)
```
