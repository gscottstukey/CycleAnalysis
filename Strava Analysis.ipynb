```python
def get_hill_features(hill, StravaAnalyzer, ride_idx):
    hill_alt = StravaAnalyzer.alt[hill['start']:hill['stop'], ride_idx]
    hill_time = StravaAnalyzer.time[hill['start']:hill['stop'], ride_idx]
    hill_distance = StravaAnalyzer.dist[hill['start']:hill['stop'], ride_idx]
    time_elapsed = (hill_time - hill_time[0])#[:, np.newaxis]
    distance_elapsed = (hill_distance - hill_distance[0])#[:, np.newaxis]
    climb_elapsed = (hill_alt - hill_alt[0])#[:, np.newaxis]
#     print hill_alt[-20:] - hill_alt[0]
    grade_smooth = smooth(StravaAnalyzer.grade[:, ride_idx])[hill['start']:hill['stop']]
    velocity_smooth = smooth(StravaAnalyzer.vel[:, ride_idx])[hill['start']:hill['stop']]
#     print grade_smooth.shape, time_elapsed.shape, distance_elapsed.shape,
#     print climb_elapsed.shape, velocity_smooth.shape
#     print hill['start'], hill['stop']
    df = pd.DataFrame({'hill_grade_smooth': grade_smooth,
                      'time_elapsed': time_elapsed,
                      'distance_elapsed': distance_elapsed,
                      'climb_elapsed': climb_elapsed,
                      'velocity_smooth': velocity_smooth},
                     index=np.arange(grade_smooth.shape[0]))
    

    return df

def hills_df(StravaAnalyzer, ride_idx):
    hills = StravaAnalyzer.hill_analysis(ride_idx)
    
    for i, hill in enumerate(hills):
        hill_df = get_hill_features(hill, StravaAnalyzer, ride_idx)
        if i == 0:
            hills_df = hill_df
        else:
            hills_df = hills_df.append(hill_df, ignore_index=True)
    return hills_df
    
```
