```python
def velocity_to_time(v, d):
    t = np.zeros(v.shape[0])
    dx = np.diff(d)
    for i in xrange(1, v.shape[0]):
        t[i] = dx[i-1] / ((v[i-1]+v[i]) / 2.)
        
    return np.cumsum(t)

def vel_to_time(v, d):
    dx = np.diff(d)
    v_avg = (v + np.roll(v, -1)) / 2.
    t = dx / v_avg[:-1]
    t = np.append(0, t)
    return np.cumsum(t)
    
```
