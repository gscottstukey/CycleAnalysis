```python
import gpxpy
import matplotlib.pyplot as plt
import pandas as pd
import collections
import numpy as np
from StravaEffort import StravaActivity
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor, AdaBoostRegressor
from StravaDB import StravaDB
from StravaUser import StravaUser
from SignalProc import smooth, vel_to_time
from sklearn.cross_validation import cross_val_score, train_test_split
import pickle
from sklearn.grid_search import GridSearchCV
%matplotlib inline
```

```python
user = StravaUser(4478600, get_streams=True, get_routes=False)
```

```python
indices = np.arange(len(user.activities))
train_indices = np.random.choice(indices, size=int(len(user.activities)*0.75), replace=False)
test_indices = np.setdiff1d(indices, train_indices)
```

```python
df = user.make_df(train_indices)
if 'predicted_time' in df.columns:
    df.pop('predicted_velocity')
    df.pop('predicted_time')
df.info()
```

```python
y = df['velocity'].values
cols = df.columns
X = df[cols[np.where(cols!='velocity')]].values
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3)
```

###Gradient Boosting

```python
model = GradientBoostingRegressor(max_depth=3, min_samples_leaf=1000)
model.fit(X_train, y_train)
```

```python
sorted(zip(model.feature_importances_, df.columns), key=lambda x: x[0])
```

```python
errors = []
normed_errors = []
for idx, a in enumerate(user.activities):
    y = a.df.velocity
    if 'predicted_time' in a.df.columns:
        a.df.pop('predicted_velocity')
        a.df.pop('predicted_time')
    a.predict(model)
    pred_time = a.df.predicted_time.iloc[-1]
    actual = a.df.time.iloc[-1] - a.df.time.iloc[0]
    error = (pred_time - actual) / actual
    norm_error = (1./a.df.shape[0])*np.linalg.norm(a.df.predicted_velocity - y)
    errors.append(error)
    normed_errors.append(norm_error)
    print '%0.3f, %0.3f, %s, %s' % (error*100, norm_error, a, idx)
print test_indices
```

```python
fig, axs = plt.subplots(1,2, figsize=(15, 3))
axs[0].hist(normed_errors)
axs[1].hist(errors)
```

###Random Forest

```python
from time import time
from operator import itemgetter
from scipy.stats import randint as sp_randint

# from sklearn.grid_search import GridSearchCV, RandomizedSearchCV
from sklearn.datasets import load_digits
def report(grid_scores, n_top=3):
    top_scores = sorted(grid_scores, key=itemgetter(1), reverse=True)[:n_top]
    for i, score in enumerate(top_scores):
        print("Model with rank: {0}".format(i + 1))
        print("Mean validation score: {0:.3f} (std: {1:.3f})".format(
              score.mean_validation_score,
              np.std(score.cv_validation_scores)))
        print("Parameters: {0}".format(score.parameters))
        print("")
params = {'max_depth': [3, 6, 8],
          'max_features': [2, 6, 8]}
params = {'min_samples_split': [2, 1000],
          'min_samples_leaf': [1, 1000]}
grid_search = GridSearchCV(GradientBoostingRegressor(), param_grid=params, n_jobs=-1)
start = time()
grid_search.fit(X_train, y_train)

print("GridSearchCV took %.2f seconds for %d candidate parameter settings."
      % (time() - start, len(grid_search.grid_scores_)))
report(grid_search.grid_scores_)
```
