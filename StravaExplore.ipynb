```python
def predict(df, window_len=wlen):
    pred = np.zeros(df.shape[0])
    prev_vel = np.ones(window_len)
    for i, row in df.iterrows():
        for j in xrange(window_len // 2):
            row['velocity_' + str(-(j+1))] = prev_vel[j]
        time_int = model.predict(row[np.newaxis, :])[0]
#         print time_int
        vel = row['dist_int'] / time_int
#         print vel
        prev_vel = np.roll(prev_vel, 1)
        prev_vel[0] = vel
        pred[i] = time_int
    return pred
def predict_activity(a):
    adf = a.make_df()

#     ycol = 'velocity'
#     adf = adf[cols + [ycol]]
#     adf = sliding_window(adf, slide_cols, ycol)
#     adf.fillna(0, inplace=True)
    adf.pop('velocity')
    y = adf.pop('time_int')
    X = adf.values
    pred = predict(adf)
#     pred = model.predict(X)
    pred_time = np.sum(pred)
#     print adf.head(10)
    return pred, y, pred_time, a.moving_time
```
